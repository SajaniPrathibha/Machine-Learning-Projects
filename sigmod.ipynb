```python
from keras.models import Sequential # feed forward training - each node is linked with it`s predecesor //keras can use prebiuld things in tenserflow
from keras.layers import Dense # nodes are fully connected across the entire NN
import numpy #numbers arrays usage library
import pandas as pd
from google.colab import files
uploaded=files.upload()
```

```python
df = pd.read_csv("diabetes.csv")  #pd.read_csv can read comma seprated values
df.head()
```

```python
model=Sequential() # commences with creating the structure of the NN, layers are connected from left to right
```

```python
model.add(Dense(12,input_dim=8,activation='relu')) # dim= dimention for 8 feature variables
model.add(Dense(20,activation='relu')) # hidden layer -01
model.add(Dense(20,activation='relu')) # hidden layer -02
model.add(Dense(22,activation='relu')) # hidden layer -03
model.add(Dense(1,activation='sigmoid')) # output layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) # configuring your back-propergation execution # configure learning process - compiler
```

```python
inputs = df.drop(['Outcome'],axis='columns') # independant dataset, excluding the dependent attibute\n", "inputs"
```

```python
target=df.Outcome # dependant attribute #expected outcome
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2) # 20% is allocated for the testing
```

```python
model.fit(X_train,y_train,epochs=100,batch_size=10) #epochs=100 mean iterations for traning and back propergation #catch size mean it devide data into 10 parts of set
```

```python
scores=model.evaluate(X_test,y_test) #evaluating using test data
```

```python
print(model.metrics_names[1],scores[1]*100)
```
