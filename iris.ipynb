```python
2from sklearn.model_selection import train_test_split
import pandas as pd #use for data manipulation and analysis
from sklearn.datasets import load_iris
iris = load_iris() # Load the Iris dataset
```

```python
dir(iris)# dir command is use to composition of data set
#give attribute in iris object
```

```python
iris.DESCR #discription of iris dataset
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names) #imports the load_iris function from scikit-learn's datasets module ( load the Iris dataset)
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
df[df.target==1].head() #this display the type on of flower versithola
```

```python
df['flower_name'] =df.target.apply(lambda x:iris.target_names[x])
df.head() #lambda use for fast execution
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'],color="orange",marker='.')
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # Split the dataset into training and testing sets
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```
